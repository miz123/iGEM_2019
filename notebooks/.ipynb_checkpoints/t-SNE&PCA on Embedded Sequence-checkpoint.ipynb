```python
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
from sklearn.cluster import KMeans
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from copy import deepcopy
```

# Using PCA then K-means

```python
#Load the data and perform PCA operation and transformation at the same time
X = np.load('/Users/Kbutterstrap/Desktop/iGEM_2019/data/seq_vectors.npy')
pca_operator = PCA(n_components = 2)
X_pca = pca_operator.fit_transform(X)
X_pca.shape
```

```python
pca_data = pd.DataFrame(X_pca, columns=['x', 'y'])
fig = plt.figure(figsize=(12,9))
ax = fig.add_subplot(1,1,1)
plt.title('Data distribution after PCA')
ax.scatter(pca_data['x'], pca_data['y'])
ax.grid()
```

```python
labels = KMeans(n_clusters=6, max_iter=1500).fit_predict(X_pca)
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=labels)
```

# PCA to lower dimension then T-SNE

```python
X = np.load('/Users/Kbutterstrap/Desktop/iGEM_2019/data/seq_vectors.npy')
pca_operator = PCA(n_components = 50)
after_pca = pca_operator.fit(X)
X_pca = pca_operator.fit_transform(X)
X_pca.shape
```

```python
#Now we visualize the relationship between PCs 
explained_variance_ratio = after_pca.explained_variance_ratio_
plt.bar(list(range(1,51)), explained_variance_ratio)
```

```python
#with default parameters of t-SNE
X_tSNE = TSNE(n_components=2, n_iter=1000, verbose=1).fit_transform(X_pca)
```
