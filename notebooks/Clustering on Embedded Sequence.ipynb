```python
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
from sklearn.cluster import KMeans
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from copy import deepcopy
from sklearn import metrics
from time import time
from sklearn.preprocessing import scale
import hdbscan
import umap
```

# Using PCA then K-means

```python
#Load the data and perform PCA operation and transformation at the same time
X = np.load('/Users/Kbutterstrap/Desktop/iGEM_2019/data/seq_vectors.npy')
pca_operator = PCA(n_components = 2)
X_pca = pca_operator.fit_transform(X)
X_pca = scale(X_pca)
X_pca.shape
```

```python
#Visualize data with only 2 dim
pca_data = pd.DataFrame(X_pca, columns=['x', 'y'])
fig = plt.figure(figsize=(12,9))
ax = fig.add_subplot(1,1,1)
plt.title('Data distribution after PCA')
ax.scatter(pca_data['x'], pca_data['y'])
ax.grid()
```

```python
#testing with sklearn
labels = KMeans(n_clusters=6, init='random', max_iter=1000).fit_predict(X_pca)
fig = plt.figure(figsize=(12,9))
ax = fig.add_subplot(1,1,1)
plt.title('Data distribution after K-Means')
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=labels)
```

# PCA to lower dimension then HDBSCAN

```python
pca_operator_new = PCA(n_components = 50)
X_pca = pca_operator.fit_transform(X)
```

```python
#choosing min_cluster_size and min_samples
cluster_labels = hdbscan.HDBSCAN(min_cluster_size=500, min_samples=10).fit_predict(X_pca)
```

```python
fig = plt.figure(figsize=(12,9))
ax = fig.add_subplot(1,1,1)
plt.title('Data distribution after HDBSCAN')
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=cluster_labels)
```

# Using UMAP then HDBSCAN

```python
embedding = umap.UMAP(
    n_neighbors=30,
    min_dist=0.0,
    n_components=50,
    random_state=42,
    verbose=True
).fit_transform(X)
plt.scatter(embedding[:, 0], embedding[:, 1])
```

```python
cluster_labels = hdbscan.HDBSCAN(
    min_samples=10,
    min_cluster_size=500,
).fit_predict(embedding)
fig = plt.figure(figsize=(12,9))
ax = fig.add_subplot(1,1,1)
plt.title('Data distribution after HDBSCAN')
plt.scatter(embedding[:, 0], embedding[:, 1], c=cluster_labels)
```
